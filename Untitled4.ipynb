```python
import os
import tarfile
import torch
import torchvision.transforms as transforms
from PIL import Image
import zipfile
import re
from pythae.models import AutoModel  # Импортируем AutoModel из pythae

# 1. Extract .tar.gz archive
def extract_model(archive_path, extract_to='.'):
    if archive_path.endswith("tar.gz"):
        with tarfile.open(archive_path, "r:gz") as tar:
            tar.extractall(path=extract_to)
        print(f'Extracted to {extract_to}')
    else:
        print('File is not a tar.gz archive')

# 2. Extract .zip archive
def extract_zip(zip_path, extract_to='.'):
    if zip_path.endswith(".zip"):  # Check if the file has a .zip extension
        with zipfile.ZipFile(zip_path, 'r') as zip_ref:
            zip_ref.extractall(extract_to)
        print(f'Extracted ZIP to {extract_to}')
    else:
        print('File is not a ZIP archive')

# 3. Load the model from Hugging Face Hub
def load_vamp_model():
    print("Loading VAMP model from Hugging Face Hub...")
    model = AutoModel.load_from_hf_hub("clementchadebec/reproduced_vamp")
    print("VAMP model loaded successfully")
    return model

# 4. Natural sort key for filenames
def natural_sort_key(s):
    """
    Splits the string into parts where numbers are converted to integers,
    ensuring proper natural sorting.
    Example: 'img10.png' -> ['img', 10, '.png']
    """
    return [int(text) if text.isdigit() else text.lower() for text in re.split('([0-9]+)', s)]

# 5. Predict digits from images in a folder using the VAMP model
def predict_digits_from_folder(folder_path, model):
    transform = transforms.Compose([
        transforms.Grayscale(num_output_channels=1),
        transforms.Resize((28, 28)),  # MNIST uses 28x28 images
        transforms.ToTensor(),
        transforms.Normalize((0.5,), (0.5,))  # Normalization for MNIST
    ])

    result = ""

    # Get list of image files and sort them naturally
    image_files = sorted(
        [f for f in os.listdir(folder_path) if f.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp'))],
        key=natural_sort_key
    )

    print("Processing files in the following order:")
    for img_file in image_files:
        print(f" - {img_file}")

    for img_file in image_files:
        img_path = os.path.join(folder_path, img_file)
        try:
            image = Image.open(img_path).convert('L')  # Convert to grayscale
            image = transform(image).unsqueeze(0)  # Add batch dimension

            with torch.no_grad():
                # Forward pass through the VAMP model
                output = model(image)

                # Extract latent representation or reconstruction
                latent_representation = output.latent_dist.mean  # Латентное представление
                pred = round(torch.sigmoid(latent_representation.mean()).item())  # Простая бинарная классификация
                result += str(pred)
        except Exception as e:
            print(f"Error processing file {img_file}: {e}")
            continue

    return result

if __name__ == "__main__":
    # Paths to files
    archive_path = 'mnist-classifier.tar.gz'
    zip_path = 'secret.zip'  # Updated file name
    extract_to = './extracted'

    # Extract the model
    extract_model(archive_path, extract_to)

    # Load the VAMP model
    vamp_model = load_vamp_model()

    # Extract the secret.zip archive
    extract_zip(zip_path, extract_to)
    secret_folder = os.path.join(extract_to, 'secret')

    # Predict digits from images in the folder
    if os.path.exists(secret_folder):
        binary_result = predict_digits_from_folder(secret_folder, vamp_model)
        print(f"Predicted binary sequence: {binary_result}")
    else:
        print("Secret folder not found after extraction.")
```

```python
def predict_digits_from_folder(folder_path, model):
    transform = transforms.Compose([
        transforms.Grayscale(num_output_channels=1),
        transforms.Resize((28, 28)),  # MNIST использует изображения 28x28
        transforms.ToTensor(),
        transforms.Normalize((0.5,), (0.5,))  # Нормализация для MNIST
    ])

    result = ""

    # Получаем список файлов и сортируем их по имени
    image_files = sorted(
        [f for f in os.listdir(folder_path) if f.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp'))],
        key=lambda x: x.lower()  # Сортировка без учета регистра
    )

    print("Processing files in the following order:")
    for img_file in image_files:
        print(f" - {img_file}")

    for img_file in image_files:
        img_path = os.path.join(folder_path, img_file)
        try:
            image = Image.open(img_path).convert('L')  # Преобразуем в градации серого
            image = transform(image).unsqueeze(0)  # Добавляем размерность батча

            with torch.no_grad():
                output = model(image)
                pred = output.argmax(dim=1).item()
                result += str(pred)
        except Exception as e:
            print(f"Error processing file {img_file}: {e}")
            continue

    return result
```
